```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O 
import matplotlib.pyplot as plt
# Input data files are available in the read-only "../input/" directory
# running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import pickle # for serializing

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# reading the datafile.
raw_data = pd.read_csv("/kaggle/input/bank-note-authentication-uci-data/BankNote_Authentication.csv")
```

```python
raw_data
```

```python
# features(X) and target(y) variable
X = raw_data.iloc[:, :-1]
y = raw_data.iloc[:, -1]
```

```python
# feature variables
X.columns
```

```python
# target variable
y.head(10)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.20)
```

```python
# Implementing classification model
from sklearn import svm
```

```python
model = svm.SVC()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Checking model performance
from sklearn.metrics import r2_score, accuracy_score
```

```python
# r2_score
r2_score(y_test, y_pred)
```

```python
# accuracy score
accuracy_score(y_test, y_pred)
```

```python
# implementing random forest classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rfc = RandomForestClassifier()
model_rfc.fit(X_train, y_train)
```

```python
# predicting
y_pred_rfc = model_rfc.predict(X_test)
y_pred_rfc
```

```python
# checking model performance
r2_score(y_test, y_pred_rfc)
```

```python
accuracy_score(y_test, y_pred_rfc)
```

```python
### Creating a pickle file
with open('svclassifier.pickle', 'wb') as file:
    pickle.dump(model, file)
    
```
